```python
sql_query = "select *, case when age >= 23 then 'Uni bitib ya da bitmek uzredir' else 'Hele de oxuyur' end as EducationInfo from students_df"
df = db.query(sql_query).to_df()
df.head(6)
```

```python
connection_string = (
    r'DRIVER={ODBC Driver 17 for SQL Server};'
    r'SERVER=LAPTOP-IVSC51O5\SQLEXPRESS;'
    r'DATABASE=AdventureWorks2022;'
    r'Trusted_Connection=yes;'
)
try:
    conn = pyodbc.connect(connection_string)
    print('Connection is Succesfull')
except Exception as e:
    print(e)
```

```python
cursor = conn.cursor()
```

```python
table_name = "NewSchema"

cursor.execute(f"""
IF OBJECT_ID('{table_name}', 'U') IS NOT NULL
    DROP TABLE {table_name};

CREATE TABLE {table_name} (
    name NVARCHAR(255),
    surname NVARCHAR(255),
    age FLOAT,
    EducationInfo TEXT
);
""")

for _, row in df.iterrows():
    cursor.execute(f"INSERT INTO {table_name} (name, surname, age, EducationInfo) VALUES (?, ?, ?, ?)", 
                   row['Name'], row['Surname'], row['Age'], row['EducationInfo'])

conn.commit()
cursor.close()
conn.close()
```

```python
# Writing SQL query for to find SaLes Person's name, surname and salesYTD
new_sqlCode = 'SELECT pp.FirstName, pp.LastName, ROUND(sp.SalesYTD, 2) AS SalesYTD FROM Person.Person pp LEFT JOIN Sales.SalesPerson sp ON pp.BusinessEntityID = sp.BusinessEntityID WHERE sp.SalesYTD IS NOT NULL'
new_df = pd.read_sql(new_sqlCode, conn)
new_df.head()
```

```python
# Creating CSV file for SalesPersons
new_df.to_excel('SalesPersons.xlsx', index=False, engine='openpyxl', sheet_name = "SalesPersons")
```
